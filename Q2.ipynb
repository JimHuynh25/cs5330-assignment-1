```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
student_data = pd.read_csv('/content/drive/MyDrive/StudentsPerformance.csv')
# Display first few rows
print(student_data.head())

# Visualization 1: Math Score Distribution
plt.figure(figsize=(8, 5))
sns.histplot(student_data["math score"], bins=10, kde=True, color="blue")
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()

# Visualization 2: Gender-wise Math Score Comparison
plt.figure(figsize=(8, 5))
sns.boxplot(x="gender", y="math score", data=student_data, palette="Set2")
plt.title("Gender-wise Math Score Distribution")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="parental level of education", y="reading score", data=student_data, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Reading Score vs Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Reading Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="test preparation course", y="writing score", data=student_data, palette="muted")
plt.title("Effect of Test Preparation on Writing Scores")
plt.xlabel("Test Preparation Course")
plt.ylabel("Writing Score")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(student_data[["math score", "reading score", "writing score"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Different Subjects")
plt.show()
```
